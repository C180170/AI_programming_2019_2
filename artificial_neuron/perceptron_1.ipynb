### 活性化関数の例

### ステップ関数（ヘビサイド関数）


\begin{eqnarray} h(y') = \begin{cases} 1 & ( y' \geq 0 ) \\ 0 & ( y' \lt 0 ) \end{cases} \end{eqnarray}

ステップ関数は、0か1しか出力できない。

```python
Image('./nn_3.jpg')
```

### シグモイド(sigmoid)関数

従来、活性化関数としてよく使われていたのがシグモイド関数である。シグモイド関数を用いると学習が停滞することがあるので、使われなくなった。

$h(y') = \displaystyle \frac{ 1 }{ 1 + \exp ( -y' )}$

```python
Image('./nn_4.jpg')
```

### ReLU(Rectified Linear Unit)関数

現在、活性化関数として最も広く用いられているのは、ReLU関数である。ReLU関数は、学習の停滞が起こりにくく、計算しやすい、という特徴がある。

\begin{eqnarray} h(y') = \begin{cases} y' & ( y' \geq 0 ) \\ 0 & ( y' \lt 0 ) \end{cases} \end{eqnarray}

```python
Image('./nn_5.jpg')
```

### シグモイド関数を描画する

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
x = np.arange(-6.0, 6.0, 0.1)
```

```python
y = sigmoid(x)

plt.plot(x, y)
```

### ReLU関数を描画する

```python
def relu(x):
    return np.maximum(0, x)
```
